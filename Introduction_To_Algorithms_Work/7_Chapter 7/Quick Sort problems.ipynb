## 7.2 - Randomised Quick Sort

```python
#Randomised Quicksort

def Printif(printyn,message):
    if printyn==1:
        print(message)

iterator,Level = 0,-1

def Partition(array,start,end,printyn=1):
    Printif(printyn,'Partition')
    i = start-1
    x = array[end]
    Printif(printyn,array)
    for j in range(start,end):
        if array[j]<=x:
            i +=1
            array[i],array[j] = array[j],array[i]
        Printif(printyn,array)
    array[i+1],array[end]=array[end],array[i+1]
    Printif(printyn,array)
    Printif(printyn,'')
    
    #Could write "return i+1, array" however the array is already updated
    return i+1

def RandomisedPartition(array,start,end,printyn=1):
    Printif(printyn,'Randomised Partition')
    i = np.random.choice(np.arange(start,end,1),1)[0]
    Printif(printyn,'Exchange values i,end = (' + str(i) + ',' + str(end) + '), A[i],A[end] = (' \
            + str(array[i]) + ',' + str(array[end]) +')')
    array[end],array[i] = array[i],array[end]
    Printif(printyn,array)
    Printif(printyn,'')
    return Partition(array,start,end,printyn=printyn)


def RandomisedQuickSort(array,start,end,printyn=1):
    
    #To help understand how it works
    global iterator,Level
    Level += 1
    Printif(printyn,'Quicksort Level = '+ str(Level) + ' array =')
    Printif(printyn,array)
    
    
    if start<end:
        print('')
        q = RandomisedPartition(array,start,end,printyn)
        print('Quicksort array after partition step')
        Printif(printyn,array)
        Printif(printyn,'')
        
        RandomisedQuickSort(array,start,q-1,printyn)
        RandomisedQuickSort(array,q+1,end,printyn)
        
        #To help understand how it works
        iterator += 1
        Printif(printyn,'Iterator = ' + str(iterator))
    
    Level += -1
    
    return array


SortThisArray = [2,8,7,1,3,5,6,4]

print(RandomisedQuickSort(SortThisArray,0,len(SortThisArray)-1))
```

### 7.2 - a Suppose that all element values are equal. What would be randomized quicksort’s running time in this case?

The partition into two subsets is done as one subset where the values are less than or equal to the pivot and one subset where the values are greater than the pivot. For this reason after each successive iteration you will get one subset with size n-1, where n was the size of the set and one subset with size 0, therefore we can model the runtime recursively as follows:<br>
<br>
$T(n) = T(n-1) + O(n)$<br>
<br>
$O(n$ in this recursion is the running time of one step of the algorithm, therefore the running time is $O(n^{2})$
