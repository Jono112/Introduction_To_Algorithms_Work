```python
import numpy as np


iterator,Level = 0,-1

def Printif(printyn,message):
    if printyn==1:
        print(message)

def HoarePartition(array,start,end,printyn=1):
    Printif(printyn,'Partition')
    x = array[start]
    i,j = start-1,end + 1
    
    Printif(printyn,'[' + ','.join([str(elem) for elem in array]) + '], i,j = (' \
    + str(i) + ',' + str(j) +') x = ' + str(x))
           
    while True:
        while True:
            j += - 1
            Printif(printyn,'i,j = ' + str(i) + ',' + str(j))
            if array[j]<=x :
                break
            else:
                pass
        while True:
            i += 1
            Printif(printyn,'i,j = ' + str(i) + ',' + str(j))
            if array[i]>=x:
                break
            else:
                pass
        if i<j:
            
            array[i], array[j] = array[j], array[i]
            Printif(printyn,'[' + ','.join([str(elem) for elem in array]) + '], i,j = (' \
                + str(i) + ',' + str(j) +') and A[i],A[j] = (' + str(array[i]) + ',' + str(array[j]) +') x = ' + str(x))
        else:
            Printif(printyn,'Condition i<j is False')
            Printif(printyn,'[' + ','.join([str(elem) for elem in array]) + '], i,j = (' \
            + str(i) + ',' + str(j) +') and A[i],A[j] = (' + str(array[i]) + ',' + str(array[j]) +') x = ' + str(x))
            break
    Printif(printyn,'')
    return j
            
    
# def Partition(array,start,end,printyn=1):
#     Printif(printyn,'Partition')
#     i = start-1
#     x = array[end]
#     Printif(printyn,array)
#     for j in range(start,end):
#         if array[j]<=x:
#             i +=1
#             array[i],array[j] = array[j],array[i]
#         Printif(printyn,array)
#     array[i+1],array[end]=array[end],array[i+1]
#     Printif(printyn,array)
#     Printif(printyn,'')
    
#     #Could write "return i+1, array" however the array is already updated
#     return i+1


def QuickSort(array,start,end,printyn=1):
    
    #To help understand how it works
    global iterator,Level
    Level += 1
    Printif(printyn,'Quicksort Level = '+ str(Level) + ' array =')
    Printif(printyn,array)
    
    
    if start<end:
        print('')
        q = HoarePartition(array,start,end,printyn)
        print('Quicksort array after partition step')
        Printif(printyn,array)
        Printif(printyn,'')
        
        QuickSort(array,start,q-1,printyn)
        QuickSort(array,q+1,end,printyn)
        
        #To help understand how it works
        iterator += 1
        Printif(printyn,'Iterator = ' + str(iterator))
    
    Level += -1
    
    return array           
            
```

## 7-1.b Prove the indices i and j are such that we never access an element of A outside the subarray

### Proof by induction

let $s,t $ and $u$ be some positions in the array between and including positions p and r

#### We have to prove $ \exists A[s] \in A[i+1..r] : A[s] \geqslant x $ and $ \exists A[t] \in A[p..j-1] : A[t] \geqslant x $ at all times in the hoare partition algorithm, because if these cases are true then we know $i$ will stop at some position $i+1$ to $ r$ and $j$ will stop in some position $p$ to $j-1$.

#### Initialisation - Case when the algorithm start with $x = A[p] < A[p+1 .. r ]$
If the array has a pivot value which is already less than all of the other values in the array, j would iterate through the array until it reached the starting position p and stop. Next i would move one position to the right and also be on value p. In this case i=j=p and therefore the hoare-partition would end. In that one iteration of the algorithm that was possible A[s]=A[p] and A[t] = A[p].

#### If this is not the case we can conclude $ \exists A[u] \in A[p+1 .. r ] : A[u] \geqslant x$.







let $q$ be some iteration value between p and r

#### At the start

$i_{0} = p-1 = -1$     (that is assuming that the indices start at 0) <br>
$j_{0} = r+1$ <br>

$x = A[p]=A[0]$

#### At iteration 1 $j$ loop
$j_{1}= j_{0}-1 = r $<br>
<br>
then<br> 
$j_{q}= j_{q-1}-1$<br>
then this continues until $A[j_{q}] \leqslant x $, this can at most occur at $j = p = 0$ <br>

#### At iteration 1 $i$ loop
$i_{1}= i_{0}+1 = 0 $<br>
<br>
then this will always stop at $i_{1}=0$ as the condition $A[i] \geqslant x$ is met as $i = 0$ and $A[i] = x$ <br> 

#### At iteration 1 final condition check
If after 1 iteration $j=p$ then the condition $i<j$ is not met and so the overall iteration loop will stop.<br>
If the condition $i<j$ is still met then this means that there must be at least one value to the right of position $i$, that is still less than $x$, meaning that upon the next iteration $i$ can loop through until this value is found and will stop there. Similarly on the next iteration j can at most get to position p.
